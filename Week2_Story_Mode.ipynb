```python
import time

story = {
    'start': {
        'text': "You find yourself in a dark forest. A path leads to four different directions.",
        'choices': {
            '1': 'left',
            '2': 'straight',
            '3': 'right',
            '4': 'end'
        }
    },
    'left': {
        'text': "You encounter a friendly gnome who gives you a magical key.",
        'choices': {
            '1': 'take_key',
            '2': 'live_alone'
        }
    },
    'take_key': {
        'text': "You take the magical key and add it to your inventory.",
        'choices': {
            '1': 'continue'
        },
        'collectibles': ['magical_key']
    },
    'continue': {
        'text': "You decide to continue your journey.",
        'choices': {}
    },
    'live_alone': {
        'text': "You decide not to continue your journey and instead choose to live alone in the wild. You survive through various challenges and eventually get old.",
        'choices': {}
    },
    'straight': {
        'text': "You stumble upon a treasure chest.",
        'choices': {
            '1': 'open_chest',
            '2': 'ignore_chest'
        }
    },
    'open_chest': {
        'text': "You open the chest and find a collection of jewels, coins, and a strength potion. What do you do?",
        'choices': {
            '1': 'take_jewels',
            '2': 'take_coins',
            '3': 'take_potion',
            '4': 'ignore_treasure'
        },
        'collectibles': ['jewels', 'coins', 'strength_potion']
    },
    'take_jewels': {
        'text': "You take the jewels and add them to your inventory.",
        'choices': {
            '1': 'continue'
        }
    },
    'take_coins': {
        'text': "You take the coins and add them to your inventory.",
        'choices': {
            '1': 'continue'
        }
    },
    'take_potion': {
        'text': "You take the strength potion and add it to your inventory.",
        'choices': {
            '1': 'continue'
        },
        'collectibles': ['strength_potion']
    },
    'ignore_treasure': {
        'text': "You decide to leave the treasure untouched and continue your journey.",
        'choices': {
            '1': 'continue'
        }
    },
    'ignore_chest': {
        'text': "You decide to ignore the treasure chest and continue your journey.",
        'choices': {
            '1': 'continue'
        }
    },
    'right': {
        'text': "You enter a cave and encounter a dragon. What do you do?",
        'choices': {
            '1': 'choose_weapon',
            '2': 'sneak',
            '3': 'run'
        }
    },
    'choose_weapon': {
        'text': "You see a selection of weapons. What weapon do you choose?",
        'choices': {
            '1': 'sword',
            '2': 'bow',
            '3': 'magic_staff'
        }
    },
    'sword': {
        'text': "You choose a sword as your weapon.",
        'choices': {
            '1': 'fight_dragon'
        },
        'collectibles': ['sword']
    },
    'bow': {
        'text': "You choose a bow as your weapon.",
        'choices': {
            '1': 'fight_dragon'
        },
        'collectibles': ['bow']
    },
    'magic_staff': {
        'text': "You choose a magic staff as your weapon.",
        'choices': {
            '1': 'fight_dragon'
        },
        'collectibles': ['magic_staff']
    },
    'fight_dragon': {
        'text': "You use your chosen weapon to fight the dragon. You emerge victorious, and the dragon drops dragon scales.",
        'choices': {
            '1': 'take_scales'
        },
        'collectibles': ['dragon_scales']
    },
    'take_scales': {
        'text': "You take the dragon scales and add them to your inventory.",
        'choices': {
            '1': 'continue'
        }
    },
    'continue': {
        'text': "You continue your journey and come across a peaceful meadow. In the distance, you see a safe and cozy home.",
        'choices': {
            '1': 'approach_home'
        }
    },
    'approach_home': {
        'text': "As you approach the home, you meet a kind lady who welcomes you. Over time, you fall in love, get married, and live a happy life together.",
        'choices': {}
    },
    'sneak': {
        'text': "You try to sneak past the dragon, but it hears you. The dragon eats you up. Game over.",
        'choices': {}
    },
    'run': {
        'text': "You run away from the dragon and escape the cave.",
        'choices': {
            '1': 'continue'
        }
    },
    'end': {
        'text': "Congratulations! You've reached the end of the game.",
        'choices': {}
    }
}

inventory = ['knife']

def display_text(text):
    for word in text.split():
        print(word, end=' ')
        time.sleep(0.1)
    print()

def make_choice(options):
    choice = input("Enter your choice: ")
    while choice not in options:
        print("Invalid choice. Try again.")
        choice = input("Enter your choice: ")
    return options[choice]

def pickup(collectibles):
    for item in collectibles:
        inventory.append(item)

def main():
    current_location = 'start'

    while current_location != 'end':
        location = story[current_location]
        display_text(location['text'])
        choices = location['choices']

        if choices:
            display_text("Choices:")
            for key, value in choices.items():
                display_text(f"{key}: {value}")

            if 'collectibles' in location:
                pickup(location['collectibles'])

            choice = make_choice(choices)

            current_location = choice
        else:
            current_location = 'end'

    print("Inventory:", inventory)

if __name__ == '__main__':
    main()
```
